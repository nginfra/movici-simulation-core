```python
import wntr
import matplotlib.pyplot as plt
import pandas as pd
```

```python
"""
A simple model connecting two reservoirs with H_R1 > H_R2
R1 -> P1 -> R2
   <- P2 <-

This model is to show the signed-ness of pipe velocity and flow rate.
The two reservoirs are connected by two pipes. One pipe is layed out in
the opposite direction. The results show that velocity is unsigned (ie.
independent of flow direction) while flow rateapi is signed (dependent on
the flow direction)
"""
wn = wntr.network.WaterNetworkModel()

# Add reservoir
wn.add_reservoir("R1", base_head=50.0)

# Add junctions
wn.add_reservoir("R2", base_head=0.0
                )

# Add pipes
wn.add_pipe(
    "P1",
    start_node_name="R1",
    end_node_name="R2",
    length=1.0,
    diameter=0.3,
    roughness=1,
    minor_loss=0.0,
)
wn.add_pipe(
    "P2",
    start_node_name="R2",
    end_node_name="R1",
    length=1.0,
    diameter=0.3,
    roughness=1,
    minor_loss=0.0,
)
sim = wntr.sim.WNTRSimulator(wn)
results = sim.run_sim()
print("velocity\n", results.link["velocity"], "\n")
print("flowrate\n", results.link["flowrate"], "\n")
```

```python
"""
A simple model connecting two reservoirs with H_R1 > H_R2
R1 -> P1 -> R2
   <- P2 <-

This model is to show the flow direction for check_valve pipes. 
The results show that a check_valve pipe only allows flow in
the forward direction: pipe P2 is restricted because the flow
tries to flow in the reverse direction
"""
wn = wntr.network.WaterNetworkModel()

wn.add_reservoir("R1", base_head=50.0, coordinates=(0.0, 0.0))
wn.add_reservoir("R2", base_head=0.0, coordinates=(1.0, 1.0))

wn.add_pipe(
    "P1",
    start_node_name="R1",
    end_node_name="R2",
    check_valve=True,
    length=1.0,
    diameter=0.3,
    roughness=1,
    minor_loss=0.0,
)
wn.add_pipe(
    "P2",
    start_node_name="R2",
    end_node_name="R1",
    check_valve=True,
    length=1.0,
    diameter=0.3,
    roughness=1,
    minor_loss=0.0,
)
sim = wntr.sim.WNTRSimulator(wn)
results = sim.run_sim()
print("velocity\n", results.link["velocity"], "\n")
print("flowrate\n", results.link["flowrate"], "\n")
```

```python
"""
A simple model connecting reservoirs to a tank
R1 -> P0 -> T0
   -> P1 -> T1
This model is to show that a pipe is sometimes shut down (status CLOSED) when
a tank is full, but not always
"""

wn = wntr.network.WaterNetworkModel()

wn.add_reservoir("R1", base_head=11)
wn.add_tank("T0", diameter=10, min_level=0, init_level=0, max_level=10)
wn.add_tank("T1", diameter=10, elevation=1, min_level=0, init_level=0, max_level=10)
wn.add_pipe(
    "P0",
    start_node_name="R1",
    end_node_name="T0",
    length=1.0,
    diameter=0.3,
    roughness=1,
)
wn.add_pipe(
    "P1",
    start_node_name="R1",
    end_node_name="T1",
    length=1.0,
    diameter=0.3,
    roughness=1,
)
DURATION = 3600 * 15
wn.options.time.duration = DURATION
wn.options.time.hydraulic_timestep = 100
wn.options.time.report_timestep = 100

sim = wntr.sim.WNTRSimulator(wn)
results = sim.run_sim()

flowrate = results.link["flowrate"]
status = results.link["status"]
head = results.node["head"]

fig, axes = plt.subplots(2)
for n, ax in enumerate(axes):
    ax = head.plot(y=[f"T{n}"], ax=ax)
    status.plot(secondary_y=True, ax=ax, y=[f"P{n}"], label=[f"P{n} status"], mark_right=False)
    ax.set_ylabel("Tank Head")
    ax.right_ax.set_ylabel("Status")
    print(
        f"final values: P{n}: {flowrate.loc[DURATION, f'P{n}']} m3/s, status: {status.loc[DURATION, f'P{n}']}. T{n}: Head {head.loc[DURATION, f'T{n}']}"
    )
```

```python
"""
A simple model connecting two reservoirs with a varying head difference
R1 -> P1 -> R2
   <- P2 <-

This model is to show that the when using check_valve pipes
the pipe status is an output variable of the model, and not just
an input variable. WNTR sets the pipe status to 0 (CLOSED) when
the check_valve suppresses backflow
"""
wn = wntr.network.WaterNetworkModel()
wn.add_pattern("r1_pat", [-50, -40, -30, -20, -10, 0, 10, 20, 30, 40, 50])
wn.add_reservoir("R1", base_head=1.0, head_pattern="r1_pat", coordinates=(0.0, 0.0))
wn.add_reservoir("R2", base_head=0.0, coordinates=(1.0, 1.0))

wn.add_pipe(
    "P1",
    start_node_name="R1",
    end_node_name="R2",
    check_valve=True,
    length=1.0,
    diameter=0.3,
    roughness=1,
    minor_loss=0.0,
)
wn.add_pipe(
    "P2",
    start_node_name="R2",
    end_node_name="R1",
    check_valve=True,
    length=1.0,
    diameter=0.3,
    roughness=1,
    minor_loss=0.0,
)
wn.options.time.duration = 10 * 3600

sim = wntr.sim.WNTRSimulator(wn)
results = sim.run_sim()
flowrate = results.link["flowrate"]
status = results.link["status"]
ax = flowrate.plot(y=["P1", "P2"], use_index=True)
status.plot(
    secondary_y=True, ax=ax, y=["P1", "P2"], label=["P1 status", "P2 status"], mark_right=False
)

ax.set_ylabel("flow rate")
ax.right_ax.set_ylabel("Status");  # semicolon to suppress output
```

```python
"""
A simple model connecting a reservoir to two tanks to show them filling up
R1 -> P1 -> T1
   -> P2 -> T2
   
This model is to show that a full tank stops flow (T1) unless the tank is
set to allow overflow (T2). However, it turns out that the WNTRSimulator
doesn't support overflow (see also
https://usepa.github.io/WNTR/apidoc/wntr.network.model.WaterNetworkModel.html#wntr.network.model.WaterNetworkModel.add_tank
)
So we can skip that attribute from our data model
"""

wn = wntr.network.WaterNetworkModel()
wn.add_reservoir("R1", base_head=5.0, coordinates=(0.0, 0.0))

wn.add_tank("T1", diameter=10, min_level=0, init_level=0, max_level=1, overflow=False)
wn.add_tank("T2", diameter=10, min_level=0, init_level=0, max_level=1, overflow=True)
wn.add_pipe(
    "P1",
    start_node_name="R1",
    end_node_name="T1",
    length=1.0,
    diameter=0.3,
    roughness=1,
    minor_loss=0.0,
)
wn.add_pipe(
    "P2",
    start_node_name="R1",
    end_node_name="T2",
    length=1.0,
    diameter=0.3,
    roughness=1,
    minor_loss=0.0,
)
wn.options.time.duration = 3600
wn.options.time.hydraulic_timestep = 100
wn.options.time.report_timestep = 100

sim = wntr.sim.WNTRSimulator(wn)
results = sim.run_sim()
flowrate = results.link["flowrate"]

head = results.node["head"]
fig, axes = plt.subplots(2)
flowrate.plot(y=["P1"], ax=axes[0])
head.plot(secondary_y=True, ax=axes[0], y=["T1"], label=["T1 head"], mark_right=False)

axes[0].set_ylabel("Pipe flow rate")
axes[0].right_ax.set_ylabel("Tank Head")

flowrate.plot(y=["P2"], ax=axes[1])
head.plot(secondary_y=True, ax=axes[1], y=["T2"], label=["T2 head"], mark_right=False)

axes[1].set_ylabel("Pipe flow rate")
axes[1].right_ax.set_ylabel("Tank Head")  # semicolon to suppress output
plt.show()
print("final status shows WNTR has turned off the pipes:")
results.link["status"].loc[3500, :]
```

```python
"""
A simple model using a pump to fill a tank
R1 -> Pu1 -> T1
   
This model is to show a POWER pump behaviour and to
show that we can change its setting and continue the simulation. It
also shows that the status is set to 0 (CLOSED) when the
tank is full
"""

wn = wntr.network.WaterNetworkModel()
wn.add_reservoir("R1", base_head=-1, coordinates=(0.0, 0.0))

wn.add_tank("T1", elevation=0, diameter=1, min_level=0, init_level=0, max_level=1, overflow=False)
wn.add_pump(
    "Pu1",
    start_node_name="R1",
    end_node_name="T1",
    pump_type="POWER",
    pump_parameter=4,
)
wn.options.time.duration = 1500
wn.options.time.hydraulic_timestep = 100
wn.options.time.report_timestep = 100

sim = wntr.sim.WNTRSimulator(wn)
results_1 = sim.run_sim()
wn.get_link("Pu1").power = 2

wn.options.time.duration += 3500
results_2 = sim.run_sim()


flowrate = pd.concat([results_1.link["flowrate"], results_2.link["flowrate"]])
head = pd.concat([results_1.node["head"], results_2.node["head"]])

ax = flowrate.plot(y=["Pu1"])
head.plot(secondary_y=True, ax=ax, y=["T1"], label=["T1 head"], mark_right=False)
ax.set_ylabel("Pipe flow rate")
ax.right_ax.set_ylabel("Tank Head")
plt.show()
print(f"Pump final status: {results_2.link['status'].loc[5000, 'Pu1']}")
```

```python
"""
A simple model using a pump to fill a tank
R1 -> Pu1 -> T1
   
This model is to show a HEAD pump behaviour. It shows
that it's HEAD curve is extrapolated when the head
falls outside the pump curve (see comment below). It 
also shows that the pump is CLOSED at the end
"""
DURATION = 2000
wn = wntr.network.WaterNetworkModel()
wn.add_reservoir("R1", base_head=-1, coordinates=(0.0, 0.0))

wn.add_tank(
    "T1", elevation=0, diameter=10, min_level=0, init_level=0, max_level=20, overflow=False
)

# pump head curve have flowrate on the x axis, and head on the y axis
# when adding a 2 point pump curve, this curve is linearly interpolated
# (and extrapolated). In this network, the initial head is 1m. The Curve
# is only defined from head 13m. To get to 1 we need to extrapolate the
# flow rate. In this case we'd expect a flow rate of 2 m3/s
wn.add_curve("Pu-curve", "HEAD", [(0, 25), (1, 13)])
wn.add_pump(
    "Pu1",
    start_node_name="R1",
    end_node_name="T1",
    pump_type="HEAD",
    pump_parameter="Pu-curve",
)
wn.options.time.duration = DURATION
wn.options.time.hydraulic_timestep = 10
wn.options.time.report_timestep = 10

sim = wntr.sim.WNTRSimulator(wn)
results = sim.run_sim()
assert results.error_code is None
head = results.node["head"]
flowrate = results.link["flowrate"]
ax = flowrate.plot(y=["Pu1"])
head.plot(secondary_y=True, ax=ax, y=["T1"], label=["T1 head"], mark_right=False)
ax.set_ylabel("Pipe flow rate")
ax.right_ax.set_ylabel("Tank Head")
plt.show()
print(f"Pump final status: {results.link['status'].loc[DURATION, 'Pu1']}")
```

```python
"""
A simple model of a tank emptying out in a reservoir
T1 -> P1 -> R1

Playing with the tank parameters reveals that for a 
Volume curve tank, the relevant properties are 
min_level, init_level, max_level, elevation and vol_curve.
The level is mapped to the volume curve to calculate the volume
in the tank. Therefore the curve should be monotonically
increasing, both in x as in y
"""
wn = wntr.network.WaterNetworkModel()
DURATION = 200
wn.add_reservoir("R1", base_head=1)

# Volume curve x=depth (level?) y=volume. Curve
# should be monotonically increasing
wn.add_curve("volume", "VOLUME", [(0, 0), (1, 1), (2, 2)])

# relevant arguments: elevation, min_level, init_level, max_level, vol_curve
# ignored arguments: diameter, overflow (also ignored for diameter tanks)
#
# min_level, init_level and max_level are checked against each other,
# and must fall in the range of the volume curve
wn.add_tank(
    "T1",
    elevation=2,
    diameter=100,
    min_level=0,
    init_level=2,
    vol_curve="volume",
    max_level=2,
    overflow=False,
)
tank = wn.get_node("T1")
wn.add_pipe(
    "P1",
    start_node_name="T1",
    end_node_name="R1",
    length=1.0,
    diameter=0.3,
    roughness=1,
)
wn.options.time.duration = DURATION
wn.options.time.hydraulic_timestep = 1
wn.options.time.report_timestep = 1

sim = wntr.sim.WNTRSimulator(wn)
print(f"[t=0] ({tank.head=!s}) = ({tank.level=!s}) + ({tank.elevation=!s})")
results = sim.run_sim()
assert results.error_code is None

flowrate = results.link["flowrate"]
status = results.link["status"]
head = results.node["head"]

print(f"[t={DURATION}] ({tank.head=!s}) = ({tank.level=!s}) + ({tank.elevation=!s})")
ax = flowrate.plot(y=["P1"], use_index=True)

# Tank head = level + elevation
head.plot(secondary_y=True, ax=ax, y=["T1"], label=["T1 head"], mark_right=False)
ax.set_ylabel("flow rate")
ax.right_ax.set_ylabel("Tank Head");  # semicolon to suppress output
```

```python
"""
A simple model, trying out different valve types

R1 -> P1 -> J1 -> V1 -> J2 -> P2 -> R2

Trying different valve types, shows that the following types
are supported in WNTR:

* PRV (Pressure Reducing Valve)
* PSV (Pressure Sustaining Valve)
* FCV (Flow Control Valve)
* TCV (Throttle Control Valve)

Unsupported valve types are PBV and GPV.

Valves are confirmed to operate in 3 statuses
(OPEN, CLOSED and ACTIVE) 

"""
import numpy as np

TIMESTEP = 3600
N_STEPS = 20

DURATION = TIMESTEP * N_STEPS
MAX_RESERVOIR_HEAD = 0.6
wn = wntr.network.WaterNetworkModel()
wn.options.time.duration = DURATION
wn.options.time.hydraulic_timestep = DURATION/N_STEPS
wn.options.time.report_timestep = DURATION/N_STEPS

wn.add_pattern("r1_pat", np.linspace(0, MAX_RESERVOIR_HEAD, num=N_STEPS+1))
wn.add_reservoir("R1", base_head=1, head_pattern="r1_pat", coordinates=(0.0, 0.0))
wn.add_reservoir("R2", base_head=0, coordinates=(1.0, 1.0))
wn.add_junction("J1", elevation=0)
wn.add_junction("J2", elevation=0)
wn.add_pipe(
    "P1",
    start_node_name="R1",
    end_node_name="J1",
    length=1.0,
    diameter=0.3,
    roughness=1,
    minor_loss=0.0,
)
wn.add_pipe(
    "P2",
    start_node_name="J2",
    end_node_name="R2",
    length=1.0,
    diameter=0.35,
    roughness=1,
    minor_loss=0.0,
)
wn.add_valve(
    "V1",
    start_node_name="J1",
    end_node_name="J2",
    valve_type="PSV",
    initial_setting=0.1,
    diameter=0.3
)

sim = wntr.sim.WNTRSimulator(wn)
results = sim.run_sim()
assert results.error_code is None

flowrate = results.link["flowrate"]
status = results.link["status"]
head = results.node["head"]

ax=head.plot(y=['J1', 'J2'], label=["Head J1", "Head J2"])
status.plot(secondary_y=True, ax=ax, y=["V1"], label=["V1 status"], mark_right=False)
ax.set_ylabel("Junction Head")
ax.right_ax.set_ylabel("Status");  # semicolon to suppress output
```

```python
"""
A simple model, trying out different valve types

R1 -> P1 -> J1 -> V1 -> J2 -> P2 -> R2

This model show cases a PRV (Pressure Reducing Valve)
"""
import numpy as np

TIMESTEP = 3600
N_STEPS = 20

DURATION = TIMESTEP * N_STEPS
MAX_RESERVOIR_HEAD = 0.6
wn = wntr.network.WaterNetworkModel()
wn.options.time.duration = DURATION
wn.options.time.hydraulic_timestep = DURATION/N_STEPS
wn.options.time.report_timestep = DURATION/N_STEPS

wn.add_pattern("r1_pat", np.linspace(-MAX_RESERVOIR_HEAD, MAX_RESERVOIR_HEAD, num=N_STEPS+1))
wn.add_reservoir("R1", base_head=1, head_pattern="r1_pat", coordinates=(0.0, 0.0))
wn.add_reservoir("R2", base_head=0, coordinates=(1.0, 1.0))
wn.add_junction("J1", elevation=0)
wn.add_junction("J2", elevation=0)
wn.add_pipe(
    "P1",
    start_node_name="R1",
    end_node_name="J1",
    length=1.0,
    diameter=0.3,
    roughness=1,
    minor_loss=0.0,
)
wn.add_pipe(
    "P2",
    start_node_name="J2",
    end_node_name="R2",
    length=1.0,
    diameter=0.35,
    roughness=1,
    minor_loss=0.0,
)
wn.add_valve(
    "V1",
    start_node_name="J1",
    end_node_name="J2",
    valve_type="PRV",
    initial_setting=0.1,
    diameter=0.3
)

sim = wntr.sim.WNTRSimulator(wn)
results = sim.run_sim()
assert results.error_code is None

flowrate = results.link["flowrate"]
status = results.link["status"]
head = results.node["head"]

fig, axes = plt.subplots(2)
flowrate.plot(y=["V1"],ax=axes[0], label=["flowrate"], use_index=True)
status.plot(secondary_y=True, ax=axes[0], y=["V1"], label=["V1 status"], mark_right=False)
axes[0].set_ylabel("Valve flow rate")
axes[0].right_ax.set_ylabel("Status");  # semicolon to suppress output

head.plot(y=['J1', 'J2'], ax=axes[1], label=["H upstream", "H downstream"])
status.plot(secondary_y=True, ax=axes[1], y=["V1"], label=["V1 status"], mark_right=False)
axes[1].set_ylabel("Junction Head")
axes[1].right_ax.set_ylabel("Status");  # semicolon to suppress output

# As can be seen from the figures, a PRV maintains a downstream pressure.
# The valve status is given according to the following rules with
# Hu being the upstream Head and Hd the downstream head. Hs is the maximum
# allowed upstream head:
# 
# * Hd > Hu -> CLOSED to prevent backflow
# * Hs > Hu > Hd -> OPEN, Maximum Hd not reached yet
# * Hu > Hs > Hd -> ACTIVE to maintain downstream Head
```

```python
"""
A simple model, trying out different valve types

R1 -> P1 -> J1 -> V1 -> J2 -> P2 -> R2

This model show cases a PSV (Pressure Sustaining Valve) through
different statuses
"""
import numpy as np
import pandas as pd

TIMESTEP = 3600
N_STEPS = 20

DURATION = TIMESTEP * N_STEPS
MAX_RESERVOIR_HEAD = 0.6
wn = wntr.network.WaterNetworkModel()
wn.options.time.duration = TIMESTEP * 15
wn.options.time.hydraulic_timestep = DURATION/N_STEPS
wn.options.time.report_timestep = DURATION/N_STEPS

wn.add_pattern("r1_pat", np.linspace(0, MAX_RESERVOIR_HEAD, num=N_STEPS+1))
wn.add_reservoir("R1", base_head=1, head_pattern="r1_pat", coordinates=(0.0, 0.0))
wn.add_reservoir("R2", base_head=0, coordinates=(1.0, 1.0))
wn.add_junction("J1", elevation=0)
wn.add_junction("J2", elevation=0)
wn.add_pipe(
    "P1",
    start_node_name="R1",
    end_node_name="J1",
    length=1.0,
    diameter=0.3,
    roughness=1,
    minor_loss=0.0,
)
wn.add_pipe(
    "P2",
    start_node_name="J2",
    end_node_name="R2",
    length=1.0,
    diameter=0.35,
    roughness=1,
    minor_loss=0.0,
)
wn.add_valve(
    "V1",
    start_node_name="J1",
    end_node_name="J2",
    valve_type="PSV",
    initial_setting=0.1,
    diameter=0.3
)
valve = wn.get_link("V1")

# first run
sim = wntr.sim.WNTRSimulator(wn)
results = sim.run_sim()
assert results.error_code is None
print(valve._internal_status, valve.initial_status, valve.status)

flowrate = results.link["flowrate"]
status = results.link["status"]
head = results.node["head"]

# interrupt the simulation and CLOSE the valve
wn.options.time.duration += TIMESTEP * 2
valve.initial_status = "CLOSED"
results = sim.run_sim()
assert results.error_code is None

print(valve._internal_status, valve.initial_status, valve.status)
flowrate = pd.concat([flowrate, results.link["flowrate"]])
status = pd.concat([status, results.link["status"]])
head = pd.concat([head, results.node["head"]])

# interrupt the simulation and reset the valve to ACTIVE
wn.options.time.duration = DURATION
valve.initial_status = 2
results = sim.run_sim()
assert results.error_code is None
print(valve._internal_status, valve.initial_status, valve.status)
flowrate = pd.concat([flowrate, results.link["flowrate"]])
status = pd.concat([status, results.link["status"]])
head = pd.concat([head, results.node["head"]])


fig, axes = plt.subplots(2)
flowrate.plot(y=["V1"],ax=axes[0], label=["flowrate"], use_index=True)
status.plot(secondary_y=True, ax=axes[0], y=["V1"], label=["V1 status"], mark_right=False)
axes[0].set_ylabel("Valve flow rate")
axes[0].right_ax.set_ylabel("Status");  # semicolon to suppress output

head.plot(y=['J1', 'J2'], ax=axes[1], label=["Head J1", "Head J2"])
status.plot(secondary_y=True, ax=axes[1], y=["V1"], label=["V1 status"], mark_right=False)
axes[1].set_ylabel("Junction Head")
axes[1].right_ax.set_ylabel("Status");  # semicolon to suppress output

# As can be seen from the figures, a PSV tries to keep the back pressure at a
# specific value (ie 0.1m). It is fully CLOSED until the back Head reaches the
# set point. Then it goes into ACTIVE status. As the source Head increases, the
# flow rate increases and the pressure drop in the downstream pipe increases,
# until that pipe becomes the limiting factor. At that point, pressure downstream
# from the valve is higher than its set point and the valve fully OPENs
# The figures also show what happens when an ACTIVE valve is set to CLOSED
# and then back to ACTIVE (around t=60000). The valve closes as expected and
# when turned back to ACTIVE is internally updated to OPEN, also as expected.
```

```python
"""
A simple model, trying out different valve types

R1 -> P1 -> J1 -> V1 -> J2 -> P2 -> R2

This model show cases a FCV (Flow Control Valve)
"""
import numpy as np

TIMESTEP = 3600
N_STEPS = 20

DURATION = TIMESTEP * N_STEPS
MAX_RESERVOIR_HEAD = 10
wn = wntr.network.WaterNetworkModel()
wn.options.time.duration = DURATION
wn.options.time.hydraulic_timestep = DURATION/N_STEPS
wn.options.time.report_timestep = DURATION/N_STEPS

wn.add_pattern("r1_pat", np.linspace(0, MAX_RESERVOIR_HEAD, num=N_STEPS+1))
wn.add_reservoir("R1", base_head=1, head_pattern="r1_pat", coordinates=(0.0, 0.0))
wn.add_reservoir("R2", base_head=0, coordinates=(1.0, 1.0))
wn.add_junction("J1", elevation=0)
wn.add_junction("J2", elevation=0)
wn.add_pipe(
    "P1",
    start_node_name="R1",
    end_node_name="J1",
    length=1.0,
    diameter=0.5,
    roughness=1,
    minor_loss=0.0,
)
wn.add_pipe(
    "P2",
    start_node_name="J2",
    end_node_name="R2",
    length=1.0,
    diameter=0.5,
    roughness=1,
    minor_loss=0.0,
)
wn.add_valve(
    "V1",
    start_node_name="J1",
    end_node_name="J2",
    valve_type="FCV",
    initial_setting=0.05,
    diameter=0.3
)

sim = wntr.sim.WNTRSimulator(wn)
results = sim.run_sim()
assert results.error_code is None

flowrate = results.link["flowrate"]
status = results.link["status"]
head = results.node["head"]

fig, axes = plt.subplots(2)
flowrate.plot(y=["V1"],ax=axes[0], label=["flowrate"], use_index=True)
status.plot(secondary_y=True, ax=axes[0], y=["V1"], label=["V1 status"], mark_right=False)
axes[0].set_ylabel("Valve flow rate")
axes[0].right_ax.set_ylabel("Status");  # semicolon to suppress output

head.plot(y=['J1', 'J2'], ax=axes[1], label=["Head J1", "Head J2"])
status.plot(secondary_y=True, ax=axes[1], y=["V1"], label=["V1 status"], mark_right=False)
axes[1].set_ylabel("Junction Head")
axes[1].right_ax.set_ylabel("Status");  # semicolon to suppress output

# A FCV limits the flow rate to a maximum value, allowing the upstream
# head to rise
```

```python
"""
A simple model, trying out different valve types

R1 -> P1 -> J1 -> V1 -> J2 -> P2 -> R2

This model show cases a TCV (Throttle Control Valve)
"""
import numpy as np

TIMESTEP = 3600
N_STEPS = 20

DURATION = TIMESTEP * N_STEPS
MAX_RESERVOIR_HEAD = 10
wn = wntr.network.WaterNetworkModel()
wn.options.time.duration = DURATION
wn.options.time.hydraulic_timestep = DURATION/N_STEPS
wn.options.time.report_timestep = DURATION/N_STEPS

wn.add_pattern("r1_pat", np.linspace(-MAX_RESERVOIR_HEAD, MAX_RESERVOIR_HEAD, num=N_STEPS+1))
wn.add_reservoir("R1", base_head=10, head_pattern="r1_pat", coordinates=(0.0, 0.0))
wn.add_reservoir("R2", base_head=0, coordinates=(1.0, 1.0))
wn.add_junction("J1", elevation=0)
wn.add_junction("J2", elevation=0)
wn.add_pipe(
    "P1",
    start_node_name="R1",
    end_node_name="J1",
    length=1.0,
    diameter=0.3,
    roughness=1,
    minor_loss=0.0,
)
wn.add_pipe(
    "P2",
    start_node_name="J2",
    end_node_name="R2",
    length=1.0,
    diameter=0.35,
    roughness=1,
    minor_loss=0.0,
)
wn.add_valve(
    "V1",
    start_node_name="J1",
    end_node_name="J2",
    valve_type="TCV",
    initial_setting=150,
    minor_loss=150,
    diameter=0.3
)

sim = wntr.sim.WNTRSimulator(wn)
results = sim.run_sim()
assert results.error_code is None

flowrate = results.link["flowrate"]
status = results.link["status"]
head = results.node["head"]

fig, axes = plt.subplots(2)
flowrate.plot(y=["V1"],ax=axes[0], label=["flowrate"], use_index=True)
status.plot(secondary_y=True, ax=axes[0], y=["V1"], label=["V1 status"], mark_right=False)
axes[0].set_ylabel("Valve flow rate")
axes[0].right_ax.set_ylabel("Status");  # semicolon to suppress output

head.plot(y=['J1', 'J2'], ax=axes[1], label=["Head J1", "Head J2"])
status.plot(secondary_y=True, ax=axes[1], y=["V1"], label=["V1 status"], mark_right=False)
axes[1].set_ylabel("Junction Head")
axes[1].right_ax.set_ylabel("Status");  # semicolon to suppress output

# a TCV is the most physical representation of a valve. It introduces a certain
# loss factor that causes the pressure drop to increase with flow rate,
# similarly to how an actual valve works when it is (partially) open. 
# It doesn't have a target pressure or flow rate, so basically always
# operates in ACTIVE status. It also allows backflow.
```

```python
"""
A simple model, trying out different valve types

R1 -> P1 -> J1 -> V1 -> J2 -> P2 -> R2

This model show cases how a TCV (Throttle Control Valve)
minor loss is ignored. In case the loss setting is lower
than its minor loss (ie. when the valve is fully open),
that setting should be ingored since it doesn't make any
sense. However, until the valve is set manually to open,
the minor loss coefficient is ignored
"""
import numpy as np
import pandas as pd

wn = wntr.network.WaterNetworkModel()

wn.add_reservoir("R1", base_head=10,  coordinates=(0.0, 0.0))
wn.add_reservoir("R2", base_head=0, coordinates=(1.0, 1.0))
wn.add_junction("J1", elevation=0)
wn.add_junction("J2", elevation=0)
wn.add_pipe(
    "P1",
    start_node_name="R1",
    end_node_name="J1",
    length=1.0,
    diameter=0.3,
    roughness=1,
    minor_loss=0.0,
)
wn.add_pipe(
    "P2",
    start_node_name="J2",
    end_node_name="R2",
    length=1.0,
    diameter=0.35,
    roughness=1,
    minor_loss=0.0,
)
wn.add_valve(
    "V1",
    start_node_name="J1",
    end_node_name="J2",
    valve_type="TCV",
    initial_setting=150,
    minor_loss=150,
    diameter=0.3
)
valve = wn.get_link("V1")
wn.options.time.duration = 100
wn.options.time.hydraulic_timestep = 10
wn.options.time.report_timestep = 10

# first part of the simulation
sim = wntr.sim.WNTRSimulator(wn)
results = sim.run_sim()
assert results.error_code is None
head = results.node["head"] 
status = results.link["status"]

# interrupt the sim and and update the setting
valve.initial_setting = 100
wn.options.time.duration = 200
results = sim.run_sim()
assert results.error_code is None
head = pd.concat([head, results.node['head']])
status = pd.concat([status, results.link['status']])

# interrupt the sim and set the status
valve.initial_status = "OPEN"
wn.options.time.duration = 200
results = sim.run_sim()
assert results.error_code is None
head = pd.concat([head, results.node['head']])
status = pd.concat([status, results.link['status']])

ax = head.plot(y=['J1', 'J2'], label=["Head J1", "Head J2"])
status.plot(secondary_y=True, ax=ax, y=["V1"], label=["V1 status"], mark_right=False)
ax.set_ylabel("Junction Head")
ax.right_ax.set_ylabel("Status");  # semicolon to suppress output

# Here we see three results
# 1) First we have a loss setting of 150, this equals the minor loss
# 2) Then we "open up" the valve further, decreasing the loss setting
# 3) Then we set the status to OPEN, now the loss setting is ignored and
#    we fall back to the minor loss (ie, equal to the setting in step (1)
```
